How would you define reinforcement learning? How is it different from
regular supervised or unsupervised learning?

RL is a technique to train agents in an environments where there are observations, actions and rewards. The agents must perform actions in order to achieve an objective, for that, the reward tells them wether they are doing right or wrong.
A great example are games. In chess there is a position, posible actions to take and the evaluation of wether an action is good or wrong is key to play, and of course, there is a goal to achieve. So an agent can be trained on policies to play this game

What is the discount factor? Can the optimal policy change change if the discount factor is modified ?

The discount factor is a measure of how much the future actions impact in the measurement of optimality of a certain action in a certain state. Of course that the policy is affected by this value, because is a way of telling the agent how much the future actions must be taken in account, so some action might be better than other based on their future posibilities

How do you measure the performance of a reinforcement learning agent?


The general answer is: By maximizing the average episodic reward return. But, there other metrics to take into account that depends on the environment and the purpose of the agent:

- Success Rate for when a task has an specific objetive, like chess or go, the performance of the agent can be expressed in terms of wether it wins or looses, fails or succeds.
- Episode Length: Sometimes the fastest the better or otherwise, a task can be to maintain some condition for longest as posible
- Stability or variance: The measure of the variability of the performance of the agent. An agent should be consistent in order to be trustable
- Generalization: How well the agent performs in unseen scenarios
- Robustness: How the agent responds to changes when there is noise, disturbances, partial failures, slightly changed environments dynamics

What is the credit assignment problem? When does it occur? How can you alleviate it?

If we always knew which step is the best for a state, training a network will only be a supervised learning task. But the reward in the task might be sparse. If you have 200 steps, the credit assignment problem is to find out which of the actions taken were good or bad after all.

To solve this problem, actions should be evaluated based on the sum of all reward that can be obtained by following them.
We must estimate how good or bad an action might be compared to other possible. This is called the action advantage, and to calculate this, many episodes should be runned and after that, normalize all the action returns. The policy gradients is an algorithm to train a neural network based on this principle.

What is the point of using a replay buffer?

The replay buffer purpose is to avoid bias by the correlation of consecutive actions taken to train in the deep Q learning algorithm. The agents tend to experience similar situations in close sets of actions. To avoid this, a buffer is needed, so the agent can be trained in a set of different experiences in the same training batch, avoiding overfitting to just some contexts.

What is an off-policy RL algorithm?

It is an algorithm that can perform actions that inquire avoiding what the policy says that should be done. This is a way of making the agent explore new situations during training and break cycles of repetitive loops that make the training ineficient. This algorithms learn the policy by following a different policy than the learned.
